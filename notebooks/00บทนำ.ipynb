```python
%pip install ipympl
%matplotlib widget
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.font_manager.fontManager.addfont("fonts/LeelawUI.ttf")
mpl.rc('font', family="Leelawadee UI")
plt.style.use('fonts/phy1.mplstyle')
```

```python
fig = plt.figure()
plt.ylabel("S")
plt.xlabel("t")
plt.title(label="เวลา t (s) และ ระยะทาง s (m)")
plt.plot(np.sin(np.linspace(0, 20, 100)), label = "ทดสอบ")
plt.legend()
plt.show()
```
