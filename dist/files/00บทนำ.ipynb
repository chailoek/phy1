# การเคลื่อนที่บนพื้นเอียง (ตัวอย่างของทฤษฎีทางฟิสิกส์)
ตัวอย่างนี้ได้อาศัย ข้อมูลจาก การทดลองใน youtube https://www.youtube.com/watch?v=ZBr8Q2ROX9s 
การทดลองนี้ทำเพื่อ ตรวจสอบ สมมุติฐานของการเคลื่อนที่ของวัตถุที่ผิวโลก สมมุติฐานคือ วัตถุจะเคลื่อนที่ด้วยอัตราเร็วไม่คงที่ กล่าวคือ
ถ้าเคลื่อนที่ออกจากผิวโลก ความเร็วจะลดลง ถ้าเคลื่อนที่เข้าหาผิวโลก อัตราเร็วจะเพิ่มขึ้น เป็นสมมุติฐานได้มาจากการสังเกตุได้จากปรากฏการณ์ในชีวิตประจำวัน 
เช่น เมื่อโยนก้อนหินขึ้นมันจะเคลื่อนที่ช้าลงเรื่อยๆ ถึงจุดหนึ่งจะหยุด และ เปลี่ยนทิศการเคลื่อนที่คือตกลงมา ระหว่างตกอัตราเร็วจะเพิ่มขึ้นเรื่อยๆ


เนื่องจากว่า อัตราเร็วของวัตถุที่เคลื่อนที่ในแนวดิ่งเปลี่ยนร็วมาก  การทดลองทำโดยใช้พื้นเอียงที่เอียงเพื่อให้อัตราการเพิ่มของอัตราเร็วไม่มากเกินไปที่จะวัดได้
การทดลองทำโดย ปล่อยลูกกลมเหล็ก ให้กลิ้งลงจากพื้นเอียงโดยกำหนดระยะทาง $s$ แตกต่างกัน 4 และ วัดเวลา $t$ ที่ใช้ในการเคลื่อนที่เป็นระยะทางเหล่านั้น
ในการวัดจำเป็นต้อง มีหน่วยสำหรับวัด ในตัวอย่างนี้หน่วยวัดความยาว และ เวลา ไม่ใช่หน่วยสากลที่นิยมใช้กัน แต่ เป็นหน่วยที่สะดวกต่อการทดลอง

### นิยามของหนึ่งหน่วยเวลา 
> เวลา 1 T = ช่วงเวลาที่ลูกตุ้มยาว 1 L แกว่ง 1 รอบ

<img src="images/00/TDef.png" alt="นิยามของ 1 L" width="40%"/>

#### วิธีวัดเวลา
ใช้ธรรมชาติการไหลของน้ำ คือ น้ำที่ไหลออกจากท่อใดท่อหนึ่งจะมีอัตราการไหลคงที่ ซึ่งเราวัดอัตราการไหลของน้ำได้งาย 
เช่น ปล่อยให้น้ำไหลลงในภาชนะที่เตรียมไว้ในช่วงเวลาหนึ่ง เช่น 10 รอบของการแกวงของลูกตุ้ม
เมื่อเรานำน้ำไป ชั่งหามวลจะได้มวลน้ำ นำมวลน้ำนี้หารด้วย 10 T จะได้อัตราการไหลของน้ำ เช่น ถ้าชั่งมวลได้ 100 กรัม
จะรู้ว่า อัตราการไหลของน้อคือ 10 กรัม ต่อ T

ถ้าเราทราบอัตราการไหลของน้ำ และ วัดมวลน้ำที่ไหลระหว่างที่วัดถุเคลื่อนที่ เราสามารถคำนวหาเวลาการเคลื่อนที่ของวัตถุได้ ตามสมการ
$$  เวลา  = \frac{มวลน้ำที่วัดได้}{อัตราการไหลของน้ำ} $$

### ทฤษฎี และ หน่วยวัด
ทฤษฎีไม่ขึ้นกับหน่วยวัด  กล่าว คือ ถ้าทฤษฎีใดถูกในหน่วยวัดหนึ่ง (เช่น หน่วยวัดที่เราได้นิยามไปแล้ว) จะถูกต้องในหน่วยวัดอื่นด้วย
หน่วยวัดจะมีผล ต่อ ค่าของปริมาณเท่านั้น จะไม่มีผลต่อความสัมพัธ์ของปริมาณ 

## ผลทดลอง และ การวิเคราะห์

ในการวิเคราะห์ จะทำโดยใช้ code ภาษา python ซึ่งเป็นภาษาที่นิยมใช้อย่างแพร่หลาย และ มีเครื่องมือที่จำเป็นสำหรับการวิเคราะห์ที่เราต้องทำ สิ่งแรกที่ต้องทำเพื่อใช้เครื่องเหล่านี้คือ ต้องนำมันขึ้นไปในหน่วยความจำ โดยใช้คำสั่งที่ตัวภาษาเตรียมไว้ให้ คือ คำสั่ง pip import สี่บรรทัดแรกของ cell ถัดไป คือการนำกลุ่มของเครื่องมือที่เรียกในภาษานี้ว่า "โมดูล" เข้ามาใช้งาน

- โมดูลหลัก matpltlib กลุ่มของเครื่่องมือที่ใช้ทำกราฟ
- โมดูย่อย pyplot ในโมดูลหลัก matpltlib
- โมดูลหลัก numpy กล่มของเครื่องมือที่ใช้ทำการคำนวณ

จะสังเกตเห็นว่าคำสั่งใน import มี คำว่า as เป็นคำที่ใช้กำหนดชื่อย่อของโมดูล เพื่อความสะสดวกในการใช้งานเครื่องมือใน โมดูล เหล่านั้น (ไม่ต้องอ้างอิงโดยใช้ชื่อเต็ม)
การใช้เครื่องมือใน โมดูล ทำโดยใช้ชื่อ ย่อ นี้ ตามด้วย จุด "." และ ชื่อของเครื่องมือ 

```python
# %pip matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(precision=1)
np.set_printoptions(floatmode="fixed")
mpl.font_manager.fontManager.addfont("fonts/LeelawUI.ttf")
mpl.rc('font', family="Leelawadee UI")
plt.style.use('fonts/phy1.mplstyle')
```

### ผลการทดลอง

```python
# กำหนดระยะทางการเคลื่อนที่  4 ระยะที่แตกต่างกัน คือ 1, 2, 3, 4 L เก็บไว้ที่ตัวแปร s_exp
s_exp = np.array([1.0, 2.0, 3.0, 4.0]) 

# ผลการวัดเวลาซึ่งวัดโดยมวลน้ำที่ไหลระหว่างที่วัดถุเคลื่อนที่ 
# เก็บไว้ที่ตัวแปร "มวลแก้วและน้ำ" ซึ่งนำไปคำนวณเป็นเวลา ในหน่วย T และเก็บไว้ที่ t_exp
มวลแก้วเปล่า = 29.3 # กรัม
มวลแก้วและน้ำ = np.array([34.2, 53.7, 62.7, 77.3]) # กรัม

มวลน้ำ = มวลแก้วและน้ํา - มวลแก้วเปล่า
อัตราการไหลฯ = 10 # กรัม ต่อ T  
t_exp = มวลน้ำ/อัตราการไหลฯ

# แสดงผลการทดลองที่บันทึกไว้
print(s_exp, "s (L)");print(t_exp, "t (T)")
```

### การวิเคราะห์
#### ทฤษฎี 
>ระยะทาง s แปรผันตรง กับ กำลังสองของเวลา t  
$$s \; \alpha \; t^2 \;\; หรือ \;\; s = c t^2 + s_0$$ 
$$ c = ค่าคงที่การแปรผัน \;(หรือ ครึ่งหนึ่งของความเร่ง\; \frac{1}{2} a)\; \;\;, s_0 = ระยะเริ่มต้น$$

#### วิเคราะโดยอาศัยกราฟของผลการวัด และ กราฟของทฤษฎี


```python
# คำนวณกำลังสองของเวลา ที่ได้จากการทดลอง (เก็บไว้ในตัวแปร t_exp) เก็บ
# ผลการคำนวณไว้ที่ tsqr_exp
tsqr_exp = t_exp**2

# ค่าคงที่การแปรผัน หรือ ความชันของเส้นตรง
c = (s_exp[2] - s_exp[1])/(tsqr_exp[2] - tsqr_exp[1])

# กำหนดช่วงเวลาสำหรับการคำนวณ คือ เริ่มจาก 0 ไปสิ้นสุดที่ 5  โดยเพิ่มที่ละ 
# 0.05 (5/100) เก็บช่วงเวลา นี้ไว้ใน t_th

t_th = np.linspace(0,5,100)
s_0 = 0.85

# คำนวณระยะ s จากทฤษฎี เก็บไว้ที่ s_th
s_th = c*(t_th**2) + s_0
# ทำกราฟ ระหว่าง t^2 และ s
plt.scatter(tsqr_exp, s_exp) # ทำกราฟผลการทดลอง 
plt.plot(t_th**2, s_th) # ทำกราฟทฤษฎี

# เพิ่มข้อมูลให้แก่กราฟ
plt.title("กราฟที่ 1\nความสัมพันธ์ระหว่าง t^2 และ s ")
plt.legend(["ผลการทดลอง", f"ทฤษฎี (ความชัน={c:.1f} จุดตัดแกน y ={s_0})"])
plt.xlabel("t^2 (T^2)");plt.ylabel("s (L)")
plt.axis([0,25,0,5]);
```

```python
# ทำกราฟ ระหว่าง t และ s
plt.scatter(t_exp, s_exp)
plt.plot(t_th, s_th  )
plt.title("กราฟที่ 2\nความสัมพันธ์ระหว่าง t และ s ")
plt.legend(["ผลการทดลอง", "ทฤษฎี"])
plt.xlabel("t (T)"); plt.ylabel("s (L)");
plt.axis([0,5,0,5]);
```

## ทำการทดลองเพื่อตรวจสอบสมมุติฐานอีกครั้ง

```python
def s(t): return c*t**2 + s_0
t = np.array([1.0, 2.0, 3.0, 4.0])
print(t, " t (T)")
print(s(t), " s (L)")
```
